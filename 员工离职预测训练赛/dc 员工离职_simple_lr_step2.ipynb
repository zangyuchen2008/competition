```python
def encode_onehot(df,column_name):
    feature_df=pd.get_dummies(df[column_name], prefix=column_name)
    all = pd.concat([df.drop([column_name], axis=1),feature_df], axis=1)
    return all
```

```python
for i in cat_col:
    data = encode_onehot(data,i)
```

```python
data.head()
```

```python
data.drop(['EmployeeNumber'], axis=1, inplace=True)
# 删除唯一值
```

```python
data.head()
```

```python
feats = [i for i in data.columns if i not in ['Label','ID']]
```

```python
import numpy as np

max_min_scaler = lambda x : (x-np.min(x))/(np.max(x)-np.min(x))
for i in feats:
    data[i]=data[[i]].apply(max_min_scaler)
```

```python
data.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
train = data[data['Label'] !=-1][feats]
train=train.fillna(0)
label = data[data['Label'] !=-1]['Label']
test = data[data['Label'] ==-1][feats]
test = test.fillna(0)
model = LogisticRegression()
model.fit(train,label)
```

```python
p =model.predict_proba(test[feats])[:,1]
```

```python
test['result']=p
```

#### 直接得到单模结果

```python
test['ID']=data[data['Label'] ==-1]['ID']
test[['ID','result']].to_csv(path+'output/lr_proba.csv',index=False)
```

```python

z =model.predict(test[feats])
test['Label']=z
test[['ID','Label']].to_csv(path+'output/lr_submission.csv',index=False)
```

使用网络搜索选择最优参数

```python
from sklearn.model_selection import GridSearchCV
lg=LogisticRegression()
clf=GridSearchCV(lg,param_grid=[{'C':np.arange(0.001,0.05,0.001)}],cv=5)
clf.fit(train,label)
best_model=clf.best_estimator_
clf.best_params_
```

```python
del test['Label']
test['Label']=best_model.predict(test[feats])
test[['ID','Label']].to_csv(path+'output/grid_lrsubmission.csv',index=False)
```
